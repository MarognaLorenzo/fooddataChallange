```python
import pandas as pd
import os
```

```python
filenames=["foodsavingleuven/places.csv",
           "foodsavingleuven/members.csv"	,
           "foodsavingleuven/activities.csv",
           "foodsavingleuven/activities_participants.csv",
            "foodsavingleuven/feedback.csv"]
```

```python
places_df=pd.read_csv("../" + filenames[0],index_col=0)
places_df
```

```python
activities_df=pd.read_csv("../" + filenames[2],index_col=0)
activities_df
```

```python
from datetime import datetime, timedelta, timezone

# Convert date_start to datetime
activities_df['date_start'] = pd.to_datetime(activities_df['date_start'], utc=True)

# Define the date range
end_date = datetime(2024, 11, 1).replace(tzinfo=timezone.utc)
start_date = (end_date - timedelta(days=6*30)).replace(tzinfo=timezone.utc)

# Filter activities within the last 6 months
filtered_activities = activities_df[(activities_df['date_start'] >= start_date) & (activities_df['date_start'] <= end_date)]

# Group by placeid and count activities
activities_count_6m = filtered_activities.groupby('placeid').size().reset_index(name='activity_count')

# Merge with name from places_df
activities_count_6m = activities_count_6m.merge(places_df[['placeid', 'name']], on='placeid')

print(activities_count_6m)
```

```python
#Total activities last 6 months
activities_count_6m['activity_count'].sum()
```

```python
from datetime import datetime, timedelta, timezone

# Convert date_start to datetime
activities_df['date_start'] = pd.to_datetime(activities_df['date_start'], utc=True)

# Define the date range
end_date = datetime(2024, 11, 20).replace(tzinfo=timezone.utc)
start_date = (end_date - timedelta(days=3*30)).replace(tzinfo=timezone.utc)

# Filter activities within the last 6 months
filtered_activities = activities_df[(activities_df['date_start'] >= start_date) & (activities_df['date_start'] <= end_date)]

# Group by placeid and count activities
activities_count_3m = filtered_activities.groupby('placeid').size().reset_index(name='activity_count')

# Merge with name from places_df
activities_count_3m = activities_count_3m.merge(places_df[['placeid', 'name']], on='placeid')

print(activities_count_3m)
```

```python
#Total activities last 3 months
activities_count_3m['activity_count'].sum()
```

```python
# Plot Pareto chart of activities in last 6 months
activities_count_6m['placeid_name'] = activities_count_6m['placeid'].astype(str) + '-' + activities_count_6m['name']
activities_count_6m = activities_count_6m.sort_values('activity_count', ascending=False)
activities_count_6m['cumulative_activity_count'] = activities_count_6m['activity_count'].cumsum()
activities_count_6m['cumulative_activity_percentage'] = 100 * activities_count_6m['cumulative_activity_count'] / activities_count_6m['activity_count'].sum()

import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 8))

ax1.bar(activities_count_6m['placeid_name'], activities_count_6m['activity_count'], color='C0')
ax1.set_xlabel('Place')
ax1.set_ylabel('Activity Count', color='C0')
ax1.tick_params(axis='y', labelcolor='C0', labelsize=10)
ax1.set_xticklabels(activities_count_6m['placeid_name'], rotation=45)

ax2 = ax1.twinx()
ax2.plot(activities_count_6m['placeid_name'], activities_count_6m['cumulative_activity_percentage'], color='C1', marker='D', ms=7)
ax2.set_ylabel('Cumulative Percentage', color='C1')
ax2.tick_params(axis='y', labelcolor='C1')
ax2.axhline(50, color='r', linestyle='--')

plt.title('Amount of Activities in Last 6 Months')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
