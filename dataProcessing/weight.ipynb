```python

def getAmtActivities():
    df = (activities_df
    .merge(places_df, on="placeid", how="left")
    .merge(feedback_df, on="activity_id", how="left")
    )
    print(len(df))
    df.head()

    activity_counts = df['placeid'].value_counts()
    activity_counts.reset_index()
    return activity_counts
```

```python
def getMeans():
    df = (activities_df
    .merge(places_df, on="placeid", how="left")
    .merge(feedback_df, on="activity_id", how="left")
    )
    print(len(df))
    df.head()

    return df.groupby("placeid")["weight"].mean()
```

```python
def getNames():
    return places_df.set_index('placeid')['name'].sort_index()
```
