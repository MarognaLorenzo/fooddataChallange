```python
import pandas as pd

# Read the CSV file
place_info_df = pd.read_csv('placeOverall.csv', index_col=0)

# Display the first few rows of the dataframe
place_info_df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x = place_info_df['weight_avg'], y = place_info_df['grade'], s=place_info_df['amt'])
plt.xlabel('Weight Avg')
plt.ylabel('Grade')
plt.title('Correlation between Grade and Weight Avg')
plt.show()
```
